```python
#Department with highest average performance rating
query5 = """
SELECT Department, ROUND(AVG(PerformanceRating),2) AS AvgPerformance
FROM employees
GROUP BY Department
ORDER BY AvgPerformance DESC
LIMIT 1
"""
pd.read_sql(query5, conn)
```

```python
# job roles that have the highest attrition rate
# Why itâ€™s useful: HR can focus retention efforts on the most at risk roles. 
query6 = """
SELECT JobRole,
       ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS AttritionRate
FROM employees
GROUP BY JobRole
ORDER BY AttritionRate DESC
"""
pd.read_sql(query6, conn)
```

```python
# Plot horizontal barattrition_by_role = df.groupby('JobRole')['Attrition'].value_counts(normalize=True).unstack()['Yes'] 
attrition_by_role = df.groupby('JobRole')['Attrition'].value_counts(normalize=True).unstack()['Yes'] 
attrition_df = attrition_by_role.reset_index()
attrition_df.columns = ['JobRole', 'AttritionRate']
attrition_df = attrition_df.sort_values('AttritionRate', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='AttritionRate', y='JobRole', data=attrition_df, color='purple')
plt.title("Attrition Rate by Job Role", fontsize=16)
plt.xlabel("Attrition Rate")
plt.ylabel("Job Role")
plt.xlim(0, 1)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))  # show as %
plt.show()
```

Insight: The highest attrition rates are seen in Sales Representatives (39.8%), Laboratory Technicians (23.9%), and Human Resources (23.1%), indicating these roles are most at risk of turnover. In contrast, senior roles such as Managers, Research Directors, and Healthcare Representatives have the lowest attrition rates, suggesting more stability among experienced employees. This highlights the need for HR to focus retention efforts on frontline and operational roles.

```python
# This query calculates attrition rates for employees who work overtime vs those who don't.
# It's important because it helps HR identify if overtime is linked to higher employee turnover.
query7 = """
SELECT OverTime,
       ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*),2) AS AttritionRate
FROM employees
GROUP BY OverTime
"""
pd.read_sql(query7, conn)
```

```python
# Attrition Rate by Job Satisfaction Level
#Why: Helps HR target retention strategies for dissatisfied employees.
query8 = """
SELECT JobSatisfaction,
       ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS AttritionRate
FROM employees
GROUP BY JobSatisfaction
ORDER BY JobSatisfaction
"""

pd.read_sql(query8, conn)
```

Insight: Employees with lower job satisfaction (level 1) have the highest attrition rate (22.8%), while those with higher satisfaction (level 4) have the lowest attrition rate (11.3%). This shows that improving job satisfaction can help reduce employee turnover

```python
# This query calculates the attrition rate for each age group.
# It's useful for HR to see which age groups are more likely to leave the company.

query9 = """
SELECT Age,
       ROUND(SUM(CASE WHEN Attrition='Yes' THEN 1 ELSE 0 END)*100.0/COUNT(*), 2) AS AttritionRate
FROM employees
GROUP BY Age
ORDER BY Age
"""

pd.read_sql(query9, conn)
```
